```python
def normalize_adj(adj):
    """Symmetrically normalize adjacency matrix."""
    adj = sp.coo_matrix(adj)
    rowsum = np.array(adj.sum(1))
    d_inv_sqrt = np.power(rowsum, -0.5).flatten()
    d_inv_sqrt[np.isinf(d_inv_sqrt)] = 0.
    d_mat_inv_sqrt = sp.diags(d_inv_sqrt)
    return adj.dot(d_mat_inv_sqrt).transpose().dot(d_mat_inv_sqrt).toarray()


def preprocess_adj(adj):
    """Preprocessing of adjacency matrix for simple GCN model and conversion to tuple representation."""
    adj_normalized = normalize_adj(adj + sp.eye(adj.shape[0]))
    return adj_normalized
```

```python
def calculate_b(n):
    alpha = 0.7*n/math.log(n)
    beta = math.pow(math.sqrt(alpha)-1.6, 2)
    b = beta*math.log(n)/n
    return b
```

```python
calculate_b(500)
```

```python
def generate_graph(num_nodes, num_graphs=5, a=0.7, threshold=1.6):
    dict_graph = {}
    for i in tqdm(range(num_graphs)):
        alpha = a*num_nodes/math.log(num_nodes)
        beta = math.pow(math.sqrt(alpha)-threshold, 2)
        b = beta*math.log(num_nodes)/num_nodes
        probs = [[a,b],[b,a]]
        labels = [0]*int(num_nodes/2)
        labels.extend([1]*int(num_nodes/2))
        sizes = [int(num_nodes/2), int(num_nodes/2)]
        
        g = nx.stochastic_block_model(sizes, probs, seed=None)
        A = nx.adjacency_matrix(g)
        A = np.asarray(A.todense())
        A = preprocess_adj(A)

        dict_graph[i] = {'num_nodes': num_nodes,
               'num_communities': 2,
               'p_matrix': probs,
               'adj': A,
               'labels': labels}
    pickle.dump(dict_graph, open('./data_new/data/{}.pkl'.format(str(num_nodes)+'_'+str(threshold)), "wb"))
```

```python
generate_graph(500, num_graphs=2, threshold=1.6)
```
